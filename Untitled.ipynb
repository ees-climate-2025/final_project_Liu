```python
zstore_o2 = o2_data.zstore.values[-1]
zstore_o2sat = o2sat_data.zstore.values[-1]
zstore_o2
```

```python
# create a mutable-mapping-style interface to the store
mapper_o2 = gcs.get_mapper(zstore_o2)
mapper_o2sat = gcs.get_mapper(zstore_o2sat)

# open the dataset using xarray and zarr
o2_data = xr.open_zarr(mapper_o2, consolidated=True,use_cftime=True)
o2 = o2_data.o2
lon = o2_data.lon
lat = o2_data.lat
time = o2_data.time
lev = o2_data.lev

o2sat = xr.open_zarr(mapper_o2sat, consolidated=True,use_cftime=True).o2sat
o2
```

```python
from cmip_basins.basins import generate_basin_codes

grid = xr.Dataset()
grid["lon"] = xr.DataArray(lon, dims=("lon"))
grid["lat"] = xr.DataArray(lat, dims=("lat"))

codes = generate_basin_codes(grid, lon="lon", lat="lat", style="cmip6")
basin_mapping = {
    "Southern Ocean": 1,
    "Atlantic": 2,
    "Pacific": 3,
    "Arctic" : 4,
    "Indian": 5
}
```

```python
fig, ax = plt.subplots(2,3 ,figsize=(10,8), dpi=300)
for i, (basin, id) in enumerate(basin_mapping.items()):
    basin_o2 = o2.where(codes == id)
    basin_o2.isel(time=0, lev=0).plot(ax=ax[i])

plt.style.use("dark_background")

fig, ax = plt.subplots(1,5 ,figsize=(16,8), dpi=300, sharey=True)

for i, (basin, id) in enumerate(basin_mapping.items()):
    
    basin_o2 = o2.where(codes == id)
    basin_o2_profile = basin_o2.mean(dim=['time','lon','lat'])

    basin_o2_profile.plot(ax=ax[i], y='lev',yincrease=False, label=basin, color='red',linewidth=2)

    ax[i].set_title(basin)
    plt.suptitle('Dissolve Oxygen Mean Profiles over Basins')
```
